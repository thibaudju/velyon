```python
#import des differents tables depuis local, a reflechir pour les sortir du bucket
import pandas as pd

df_2011 = pd.read_csv('/content/caracteristiques_2011.csv', delimiter=',', encoding= 'latin1')
df_2012 = pd.read_csv('/content/caracteristiques_2012.csv', delimiter=',', encoding= 'latin1')
df_2013 = pd.read_csv('/content/caracteristiques_2013.csv', delimiter=',', encoding= 'latin1')
df_2014 = pd.read_csv('/content/caracteristiques_2014.csv', delimiter=',', encoding= 'latin1')
df_2015 = pd.read_csv('/content/caracteristiques_2015.csv', delimiter=',', encoding= 'latin1')
df_2016 = pd.read_csv('/content/caracteristiques_2016.csv', delimiter=',', encoding= 'latin1')
df_2017 = pd.read_csv('/content/caracteristiques-2017.csv', delimiter=',', encoding= 'latin1')
df_2018 = pd.read_csv('/content/caracteristiques-2018.csv', delimiter=',', encoding= 'latin1')
df_2019 = pd.read_csv('/content/caracteristiques-2019.csv', delimiter=';', encoding= 'latin1')
df_2020 = pd.read_csv('/content/caracteristiques-2020.csv', delimiter=';', encoding= 'latin1')
df_2021 = pd.read_csv('/content/caracteristiques-2021.csv', delimiter=';', encoding= 'latin1')
```

```python
#concatenation des tab
df_caracteristiques = pd.concat([df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021])
df_caracteristiques
```

```python
#modification colonne an

import pandas as pd

# Modifier la colonne en utilisant les 4 premiers chiffres de la colonne précédente
df_caracteristiques['an'] = df_caracteristiques['Num_Acc'].astype(str).str[:4].astype(int)

# Afficher le DataFrame avec la colonne modifiée
df_caracteristiques
```

```python
#rename
df_caracteristiques_all = df_caracteristiques
df_caracteristiques_all
```
