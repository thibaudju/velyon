```python
#Passer d'un fichier GEOJSON à un fichier CSV sans perdre les données géographiques
import geopandas as gpd
import io 
import pandas as pd

url = "/content/pvo_patrimoine_voirie.pvostationnementvelo.geojson"

stationnement = gpd.read_file(url)
stationnement_df = pd.DataFrame(stationnement)
stationnement_df.to_csv('stationnement.csv', index=True)
```
