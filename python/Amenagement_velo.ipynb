```python
#Passer d'un fichier GEOJSON à un fichier CSV sans perdre les données géographiques
import geopandas as gpd
import io 
import pandas as pd

url = "/content/pvo_patrimoine_voirie.pvoamenagementcyclable.geojson"

amenagement = gpd.read_file(url)
amenagement_df = pd.DataFrame(amenagement)
amenagement_df.to_csv('amenagement.csv', index=True)
```

```python
#Récupérer les informations pour chaque colonne du tableau (nombre de colonnes avec des lignes nulles notamment)
amenagement_df.info()
```

```python
#Convertir la colonne "anneelivraison" en datetime (mais on n'a que l'année à la base et là on obtient le jour et le mois)
amenagement_df['anneelivraison'] = pd.to_datetime(amenagement_df['anneelivraison'],format = '%Y')
amenagement_df.info()
```
