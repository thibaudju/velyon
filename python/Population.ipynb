```python
#Passer d'un fichier GEOJSON à un fichier CSV sans perdre les données géographiques
import geopandas as gpd
import io 
import pandas as pd

url = "/content/ter_territoire.terpopulation_latest.geojson"

population = gpd.read_file(url)
population_df = pd.DataFrame(population)
population_df.to_csv('population.csv', index=True)
```
